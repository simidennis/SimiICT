# Check whether all the zones generate the same amount of sales.


```python
H0: Amount of sales in all zone are same
H1: Amount of sales in all zone are  different
```

```python
import scipy.stats as scp
```

```python
f_value , p_value = scp.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'],data['Zone - D'])
print('The f value is:',f_value)
print('The p value is:',p_value)
```

```python
P value < .05 we reject the null hypothesis.
Amount of sales in different zones are different
```
