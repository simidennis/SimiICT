# What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
dfresult3=df[['race/ethnicity','gender','math score']]
sns.set()  # use Seaborn styles
dfresult3.pivot_table('math score', index='race/ethnicity', columns='gender', aggfunc='sum').plot()

plt.ylabel('scoring variation -gender base');
```

```python
dfresult3=df[['race/ethnicity','gender','reading score']]
sns.set()  # use Seaborn styles
dfresult3.pivot_table('reading score', index='race/ethnicity', columns='gender', aggfunc='sum').plot()

plt.ylabel('scoring variation -gender base');
```

```python
dfresult3=df[['race/ethnicity','gender','reading score']]
sns.set()  # use Seaborn styles
dfresult3.pivot_table('reading score', index='race/ethnicity', columns='gender', aggfunc='sum').plot()

plt.ylabel('scoring variation -gender base');
```

```python
dfresult3=df[['race/ethnicity','gender','writing score']]
sns.set()  # use Seaborn styles
dfresult3.pivot_table('writing score', index='race/ethnicity', columns='gender', aggfunc='sum').plot()

plt.ylabel('scoring variation -gender base');
```
